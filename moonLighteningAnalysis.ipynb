**2. Find total number of moon lightening proteins that are assigned EC number.**

```python
ec_no1_list = []
list_of_ec1 = []
list_of_dist_ec1 = []
ec_no1 = df['Fun. 1 EC'].values
for i in ec_no1:
    ec_no1_list.append(i)
for j in range(len(ec_no1_list)):
    if type(ec_no1_list[j]) is str:
        list_of_ec1.append(ec_no1_list[j].strip())
print('-----------------------------------------------------------------------------------------------------')
print(f'        Out of {len(ec_no1_list)} there are {len(list_of_ec1)} Moon Lightening Proteins that are assigned EC numbers.')
print('-----------------------------------------------------------------------------------------------------')
```

**3. Find distinct EC numbers and group repeating EC numbers**

Distinct EC numbers are less than the total number or moon lightening proteins that are assigned EC numbers. Therefore, there are some EC numbers that are repeated. The below code helps to group EC numbers according to the number of times they repeat in the database. In simple words, some EC numbers are repeated two times, so all those EC numbers will be grouped together. 

```python
count1 = []
count2 = []
count3 = []
count4 = []
count5 = []
count7 = []
count8 = []
count23 = []
count29 = []
unique = []
list_of_dist_ec1 = list(set(list_of_ec1))
for k in list_of_dist_ec1:
    if list_of_ec1.count(k) not in unique:
        unique.append(list_of_ec1.count(k))
print()
print('Some EC numbers repeat', end = ' ')
print(' times, '.join(list(map(str,sorted(unique)))), end=' times')
print()
print()
for k in list_of_dist_ec1:
    if list_of_ec1.count(k) == 1:
        count1.append(k)
    elif list_of_ec1.count(k) == 2:
        count2.append(k)
    elif list_of_ec1.count(k) == 3:
        count3.append(k)
    elif list_of_ec1.count(k) == 4:
        count4.append(k)
    elif list_of_ec1.count(k) == 5:
        count5.append(k)
    elif list_of_ec1.count(k) == 7:
        count7.append(k)
    elif list_of_ec1.count(k) == 8:
        count8.append(k)
    elif list_of_ec1.count(k) == 23:
        count23.append(k)
    else: 
        count29.append(k)
print('-------------------------------------------------------------------------------------------------------------------------------')
print()
print('List of EC numbers that donot repeat (i.e., repeat 1 time):')
print()
print(count1)
print()
print(f'Number of enzymes: {len(count1)}')
print()
print('-------------------------------------------------------------------------------------------------------------------------------')
print()
print('List of EC numbers that repeat 2 times:')
print()
print(count2)
print()
print(f'Number of enzymes: {len(count2)}')
print()
print('-------------------------------------------------------------------------------------------------------------------------------')
print()
print('List of EC numbers that repeat 3 times:')
print()
print(count3)
print()
print(f'Number of enzymes: {len(count3)}')
print()
print('-------------------------------------------------------------------------------------------------------------------------------')
print()
print('List of EC numbers that repeat 4 times:')
print()
print(count4)
print()
print(f'Number of enzymes: {len(count4)}')
print()
print('-------------------------------------------------------------------------------------------------------------------------------')
print()
print('List of EC numbers that repeat 5 times:')
print()
print(count5)
print()
print(f'Number of enzymes: {len(count5)}')
print()
print('-------------------------------------------------------------------------------------------------------------------------------')
print()
print(f'List of EC numbers that repeat 7 times:')
print()
print(count7)
print()
print(f'Number of enzymes: {len(count7)}')
print()
print('-------------------------------------------------------------------------------------------------------------------------------')
print()
print('List of EC numbers that repeat 8 times:')
print()
print(count8)
print()
print(f'Number of enzymes: {len(count8)}')
print()
print('-------------------------------------------------------------------------------------------------------------------------------')
print()
print(f'List of EC numbers that repeat 23 times:')
print()
print(count23)
print()
print(f'Number of enzymes: {len(count23)}')
print()
print('-------------------------------------------------------------------------------------------------------------------------------')
print()
print(f'List of EC numbers that repeat 29 times:')
print()
print(count29)
print()
print(f'Number of enzymes: {len(count29)}')
print()
print('-------------------------------------------------------------------------------------------------------------------------------')
print()
print('So here we can conclude that:')
print()
print(f'{len(count1)} EC numbers repeat 1 time = {len(count1)} * 1 = {len(count1)}')
print(f'{len(count2)} EC numbers repeat 2 time = {len(count2)} * 2 = {len(count2)*2}')
print(f'{len(count3)} EC numbers repeat 3 time = {len(count3)} * 3 = {len(count3)*3}')
print(f'{len(count4)} EC numbers repeat 4 time = {len(count4)} * 4 = {len(count4)*4}')
print(f'{len(count5)} EC numbers repeat 5 time = {len(count5)} * 5 = {len(count5)*5}')
print(f'{len(count7)} EC numbers repeat 7 time = {len(count7)} * 7 = {len(count7)*7}')
print(f'{len(count8)} EC numbers repeat 8 time = {len(count8)} * 8 = {len(count8)*8}')
print(f'{len(count23)} EC numbers repeat 23 time = {len(count23)} * 23 = {len(count23)*23}')
print(f'{len(count29)} EC numbers repeat 29 time = {len(count29)} * 29 = {len(count29)*29}')
print()
print(f'Thus, there are {len(count1)+len(count2)+len(count3)+len(count4)+len(count5)+len(count7)+len(count8)+len(count23)+len(count29)}', end=' ')
print('distinct EC numbers in the database.')
```

**4. Scraping the Accepted names and Alternatives for every EC number.** 

```python
import requests
import bs4
df_name = pd.read_excel('moonlighteningdatar1.xlsx') 
r1 = df_name.EC.values
```

```python
name_holder = []
for every_ec in range(len(r1)):
        website = requests.get(f'https://enzyme.expasy.org/EC/{df_name.iloc[every_ec,1]}')
        soup = bs4.BeautifulSoup(website.text, 'lxml')
        names = soup.find_all('strong')
        for every_name in names:
            name_holder.append(every_name.text)
        if 'Accepted Name' in name_holder:
            accepted_name = name_holder[1]
            df_name.at[every_ec, 'Accepted Name'] = accepted_name
            print(every_ec+1, accepted_name)
        else:
            accepted_name = '##########'
            df_name.at[every_ec, 'Accepted Name'] = accepted_name
            print(every_ec+1, accepted_name)
        if 'Alternative Name(s)' in name_holder:
            alternative_name = name_holder[name_holder.index('Alternative Name(s)')+1:name_holder.index('Reaction catalysed')]
            df_name.at[every_ec, 'Aleternative Name'] = ', '.join(alternative_name)
            print(', '.join(alternative_name))
        else:
            df_name.at[every_ec, 'Aleternative Name'] = '%%%%%'
            print('%%%%%')
        print('-----------------------------------------------------------------------------------------------------------------------------')
        name_holder.clear()
        alternative_name.clear
df_name
df_name.to_excel("moonlighteningdatar1.xlsx")
```

```python
import pandas as pd 
from selenium import webdriver 

prot_count = 1
page_count = 0
df = pd.read_excel('moonlighteningproteinwithoutec.xlsx') 
df_read = pd.read_excel('moonlighteningproteinread.xlsx')
name = df_read.Name.values 
for each_prot_name in name:
    count = 0
    df.at[prot_count, 'Protein Name'] = each_prot_name
    print(f'For {each_prot_name}:')
    chrome_browser = webdriver.Chrome()
    chrome_browser.get('https://www.uniprot.org/')
    searchbar = chrome_browser.find_element_by_id('query')
    search_button = chrome_browser.find_element_by_id('search-button')
    searchbar.clear()
    searchbar.send_keys(each_prot_name)
    search_button.click()
    swiss_prot_reviewed_button = chrome_browser.find_element_by_id('reviewed-filter')
    swiss_prot_reviewed_button.click()
    while (1):
        page_count += 1
        table = chrome_browser.find_element_by_tag_name('tbody')
        prot_id = []
        for k in range(1,len(table.get_attribute('innerHTML').split('<tr id='))):
            prot_id.append(table.get_attribute('innerHTML').split('<tr id="')[k].split('" ')[0])
        prot_name = table.find_elements_by_class_name('long')
        for each_prot_name in prot_name:
            if len(each_prot_name.get_attribute('innerHTML')) != 0:
                if 'EC' not in each_prot_name.get_attribute('innerHTML'):
                    prot_count += 1
                    prot_link = chrome_browser.find_element_by_link_text(prot_id[count])
                    df.at[prot_count, 'UniProt Id'] = prot_id[count]
                    df.at[prot_count, 'Link'] = prot_link.get_attribute('href')
                    print(prot_count, prot_id[count], prot_link.get_attribute('href'))
                count += 1  
        prot_id.clear()
        try:
            next_button = chrome_browser.find_element_by_class_name('nextPageButton')
            next_button.click()
            count = 0
            print()
            print(f'{page_count} page checked. moving to {page_count+1} page...')
            print('----------------------------------------------------------------------')
            continue
        except:
            break
    print()
    print(f'Page number {page_count} was the last page scanned. All pages checked.')
    print('Successful!!')
    chrome_browser.close()
    print('#############################################################################################')
```

### Matching aequence to find pseudo enzymes 

```python
import pandas as pd 
from selenium import webdriver
df = pd.read_excel('pseudo_enzymes_scraping.xlsx') 
```

```python
uniprot_id = df.UniProtID.values
sequence_html_extract = []
sequence_line = []
sequence_line_without_space = []
for each_uniprot_id in range(len(uniprot_id)):
    print(each_uniprot_id+1)
    chrome_browser = webdriver.Chrome()
    chrome_browser.get(f'https://www.uniprot.org/uniprot/{uniprot_id[each_uniprot_id]}')
    sequence_html = chrome_browser.find_elements_by_class_name('sequence')
    for s in sequence_html: 
        sequence_html_extract.append(s.get_attribute('innerHTML'))
    sequence_html_extract_required = sequence_html_extract[1].strip().split('<br>')
    for each_sequence_line in range(1,len(sequence_html_extract_required),2):
        sequence_line.append(sequence_html_extract_required[each_sequence_line].strip())
    for each_sub_sequence in sequence_line:
        sequence_line_without_space.append(''.join(each_sub_sequence.split()))
        ans = ''.join(sequence_line_without_space)
    df.at[each_uniprot_id, 'Sequence'] = ans
    sequence_html_extract.clear()
    sequence_line.clear()
    sequence_line_without_space.clear()
    sequence_html.clear()
    chrome_browser.close()
```
